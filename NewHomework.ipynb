В первую строку вводится число m (1 ≤ m ≤ 10e6) - максимальная масса, которую может выдержать одна лодка. Во вторую строку вводится число n (1 ≤ n ≤ 100) - количество рыбаков. В следующие N строк вводится по одному числу Ai (1 ≤  Ai ≤ m) - вес каждого путешественника.
Программа должна вывести одно число - минимальное количество лодок, необходимое для переправки всех рыбаков на противоположный берег.

**Решение задания № 3:**

````python
m=int(input('Введите максимальную грузоподъемность одной лодки в килограммах:'))
n=int(input('Введите количество рыбаков:'))
ListWeights=input('Введите  в одну строку серез пробелы вес каждого путешественника:').split()
t=sorted(ListWeights,reverse=True)
c=0
if int(t[0]) > m:
    print('Задача не имеет решения')
    exit()
while len(t):
    c += 1
    k = m - int(t.pop(0))
    for i in range(len(t)):
        if int(t[i]) <= k:
            t.pop(i)
            break
print('Минимальное количество лодок равно -',str(c))
````

В цикле, пока рыбаки не кончились:

1. Увеличиваем кол-во лодок на 1.
2. Сажаем в лодку самого тяжёлого из оставшихся рыбаков.
3. Среди оставшихся рыбаков ищем самого тяжёлого, вес которого не превышает оставшуюся после посадки первого рыбака грузоподъёмность лодки. Если нашли - сажаем его на второе место в лодке.

- Теперь создайте список факториалов чисел от получившегося ранее факториала `6`, до `1`.

В итоге, на вход программа получает например число `3`, возвращает число `6` (факториал числа `3`) и вам нужно сделать список из факториалов числа `6` в убывающем порядке. Находим факториал числа `6` - это `720`, затем от числа `5` - это `120` и так далее вплоть до `1`

То есть, результирующий список будет выглядеть в нашем примере так: <br>
`[720, 120, 24, 6, 2, 1]`

**Решение задания № 1:**

````python
def factorial(NatureNumber):
    MyOut=1
    for Number in range(NatureNumber,0,-1):
        MyOut*=Number
    return MyOut
MyList=list()
for Number in range(int(input('Введите число от которого будем вычислять факториалы по убыванию ')),0,-1):
    MyList.append(factorial(Number))
print(MyList)
````

- take_away(X) - забрать X из кассы, либо выкинуть ошибку, что не достаточно денег

````python
class cashbox:
    how_many=0
        
    def __init__(self,how_many):
        self.how_many=how_many
    def top_up(self,X):
        self.how_many=self.how_many+X
    def count_1000(self):
        print(self.how_many//1000)
    def take_away(self,X):
        if self.how_many<X:
            print('ошибка - недостаточно денег')
        else:
            self.how_many=self.how_many-X
        
        
MyCash=cashbox(2000)
MyCash.count_1000()
MyCash.top_up(5000)
MyCash.count_1000()
MyCash.take_away(10000)
MyCash.count_1000()
MyCash.take_away(3000)
MyCash.count_1000()
````

у этого класс есть методы:

- go_up() - увеличивает y на s
- go_down() - уменьшает y на s
- go_left() - уменьшает x на s
- go_right() - увеличивает y на s
- evolve() - увеличивает s на 1
- degrade() - уменьшает s на 1 или выкидывает ошибку, когда s может стать ≤ 0
- count_moves(x2, y2) - возвращает минимальное количество действий, за которое черепашка сможет добраться до x2 y2 от текущей позиции

````python
class turtle:
    s=1
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def go_up(self):
        self.y=self.y+self.s
    def go_down(self):
        self.y=self.y-self.s
    def go_left(self):
        self.x=self.x-self.s
    def go_right(self):
        self.x=self.x+self.s
    def evolve(self):
        self.s+=1
    def degrade(self):
        if(self.s==0):
            print("Ошибка. Шаг минимален и равен 0")
        else:
            self.s-=1
    def count_moves(self,x2, y2):
        if (self.x>x2):
            MyX=self.x-x2
        else:
            MyX=x2-self.x
        if (self.y>y2):
            MyY=self.y-y2
        else:
            MyY=y2-self.y
        return MyX+MyY
    
        
        
    def all_about(self):
        print(f"Координата X {self.x} Координата Y {self.y} Шаг {self.s}")
        
        
        
MyTurtle=turtle(1,1)
MyTurtle.all_about()
print(str(MyTurtle.count_moves(0,0)))
MyTurtle.degrade()
MyTurtle.all_about()
MyTurtle.degrade()
MyTurtle.all_about()
````

**Задание №3** (ответ дается текстом под кодом)

Какая сложность у этого алгоритма?

````python
n = int(input())
m = int(input())
k = 1000
for i in range(n):
	for j in range(m):
		k += 100

for i in range(m):
	for j in range(m):
		k -= 100
````

**Ответ:**O(n*m+m^2)
